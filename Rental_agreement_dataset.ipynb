## Importing Libraries

```python
import random
import pandas as pd
from faker import Faker

fake = Faker()
```

## Data generation

```python
def generate_random_date(start_date, end_date):
    return fake.date_between_dates(date_start=start_date, date_end=end_date).strftime("%B %d, %Y")

def generate_property_description():
    options = [
        "This charming three-bedroom, two-bathroom house is located in a quiet neighborhood. The house has a spacious living room with a fireplace, a dining room, a kitchen with stainless steel appliances, and a fenced-in backyard.",
        "This spacious one-bedroom, one-bathroom apartment is located in a modern building with amenities such as a fitness center, a swimming pool, and a rooftop terrace. The apartment has a large living room, a fully equipped kitchen, and a balcony with views of the city.",
        "This two-bedroom, two-bathroom condo is located in a gated community with amenities such as a tennis court, a basketball court, and a swimming pool. The condo has a spacious living room with a balcony, a fully equipped kitchen, and a master bedroom with a walk-in closet and a private bathroom.",
        "This three-bedroom, two-bathroom townhouse is located in a convenient location near shopping, restaurants, and transportation. The townhouse has a spacious living room with a fireplace, a dining room, a kitchen with stainless steel appliances, and a fenced-in backyard.",
        "This studio apartment is located in a trendy neighborhood near bars, restaurants, and shops. The apartment has a combined living and sleeping area, a kitchenette, and a bathroom.",
        "This two-bedroom, one-bathroom basement apartment is located in a single-family house. The apartment has a private entrance, a full kitchen, and a laundry room.",
        "This one-bedroom, one-bathroom loft apartment is located in a converted warehouse. The apartment has high ceilings, exposed brick walls, and a large open floor plan.",
        "This one-bedroom, one-bathroom garage apartment is located behind a single-family house. The apartment has a private entrance, a full kitchen, and a laundry room.",
        "This one-bedroom, one-bathroom ADU is located on the same property as a single-family house. The ADU has a private entrance, a full kitchen, and a laundry room.",
        "This 1,500 square foot commercial property is located in a busy retail district. The property has a large open floor plan, a storefront with windows, and a back entrance."
    ]
    return fake.random_element(options)

def generate_payment_method():
    options = [
        "Personal check",
        "Cashier's check",
        "Money order",
        "Wire transfer",
        "Online payment",
        "Automatic debit",
        "Credit card",
        "Debit card",
        "Rent collection service",
        "Cash"
    ]
    return fake.random_element(options)

def generate_utilities_included():
    options = [
        "Electricity, gas, water, sewer",
        "Internet, cable TV, phone",
        "Electricity, gas, water, trash removal",
        "Internet, cable TV, security system",
        "Gas, water, sewer, recycling",
        "Internet, phone, security system, parking",
        "Electricity, water, sewer, internet, cable TV",
        "Gas, water, trash removal, recycling, parking",
        "Internet, cable TV, security system, phone, parking",
        "Electricity, gas, water, sewer, internet, security system"
    ]
    return fake.random_element(options)

def generate_property_usage():
    options = [
        "operating a retail store selling clothing and accessories",
        "operating a restaurant serving American cuisine",
        "operating an office for a law firm",
        "operating a medical clinic",
        "operating a warehouse for a logistics company",
        "operating a manufacturing facility for a clothing company",
        "operating a fitness center",
        "operating a daycare center",
        "operating a co-working space",
        "operating a recording studio"
    ]
    return fake.random_element(options)
```

```python
rental_schema = {
    "agreement_date" : [],
    "lessor_name" : [],
    "lessor_address" : [],
    "lessee_name" : [],
    "lessee_address" : [],
    "property_address" : [],
    "property_description" : [],
    "lease_start_date" : [],
    "lease_end_date" : [],
    "lease_payment" : [],
    "due_date" : [],
    "payment_method" : [],
    "payment_address" : [],
    "security_deposit" : [],
    "number_of_days" : [],
    "utilities_included_tenant" : [],
    "utilities_included_landlord" : [],
    "notice_period" : [],
    "state" : [],
    "property_usage" : [],
    "insurance_amount" : []
}

for _ in range(1000):
    rental_schema['agreement_date'].append(generate_random_date('-3y', '+1y'))
    rental_schema['lessor_name'].append(fake.name())
    rental_schema['lessor_address'].append(fake.address())
    rental_schema['lessee_name'].append(fake.name())
    rental_schema['lessee_address'].append(fake.address())
    rental_schema['property_address'].append(fake.address())
    rental_schema['property_description'].append(generate_property_description())
    rental_schema['lease_start_date'].append(generate_random_date('-2y','+2y'))
    rental_schema['lease_end_date'].append(generate_random_date('+2y','+4y'))
    rental_schema['lease_payment'].append(random.randint(1000, 10500))
    rental_schema['due_date'].append(random.randint(1, 15))
    rental_schema['payment_method'].append(generate_payment_method())
    rental_schema['payment_address'].append(fake.address())
    rental_schema['security_deposit'].append(random.randint(500, 5000))
    rental_schema['number_of_days'].append(random.randint(1, 365))
    rental_schema['utilities_included_tenant'].append(generate_utilities_included())
    rental_schema['utilities_included_landlord'].append(generate_utilities_included())
    rental_schema['notice_period'].append(random.randint(20, 50))
    rental_schema['state'].append(fake.state())
    rental_schema['property_usage'].append(generate_property_usage())
    rental_schema['insurance_amount'].append(random.randint(1000, 5000))

rental_df = pd.DataFrame(rental_schema)

rental_df.head()
```

## Variable classification for each agreement type

Have all the variables which are common for all sub types. Additional variables can be added to specific variables.

```python
common_variables = {
    "rental_agreement" : [
        "agreement_date", "lessor_name", "lessor_address", "lessee_name", "lessee_address",
        "property_address", "property_description", "lease_start_date", "lease_end_date", "lease_payment",
        "due_date", "payment_method", "payment_address", "security_deposit", "number_of_days", "state"
    ]
}
```

```python
specific_variables = {
    "Residential Lease" : [
        #Even if sub type didn't have any additional specific variable, it will better to have empty list created for it
        #It might be useful when going for instruction/response creation for each template
    ],
    "Commercial Lease" : [
        "property_usage", "insurance_amount"
    ]
}
```

## Retrieving fields required for an agreement

```python
# Retrieving fields required for agreement

def get_fields(agreement_type, agreement_sub_type):
    return list(set(common_variables[agreement_type] + specific_variables[agreement_sub_type]))
```
